Importamos las librer√≠as que queremos utilizar

```python
import csv
import pandas as pd

from google.colab import files
upload = files.upload()

data = pd.read_csv('/content/drive/MyDrive/2023_qualitat_aire_estacions.csv') #link https://opendata-ajuntament.barcelona.cat/data/es/dataset/qualitat-aire-estacions-bcn #

filename = '2023_qualitat_aire_estacions.csv'
keys = ('Longitud', 'Latitud', 'Nom_barri')
records = []

with open(filename, 'r') as csvfile:
  reader = csv.DictReader(csvfile)
  for row in reader:
    records.append({key: row[key] for key in keys})
```

```python
print(records[0])
```

```python
for record in records:
  Longitud = record['Longitud']
  Latitud = record['Latitud']
  
  record['Longitud'] = float(Longitud)
  record['Latitud'] = float(Latitud)
```

```python
records[0]
```

```python
import folium

map = folium.Map(location=[41.382778, 2.176944], zoom_start=9)

for record in records:
  coords = (record['Latitud'], record['Longitud'])
  folium.Marker(coords, popup=record['Nom_barri']).add_to(map)
map
```
